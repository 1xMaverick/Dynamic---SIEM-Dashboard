```python
import pandas as pd

df = pd.read_csv('/kaggle/working/SIEM_data.csv')
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
```

```python
TARGET = 'SuspicionLevel_label'
features_to_drop = ['Unnamed: 0', 'SuspicionLevel', TARGET]
```

```python
X = df.drop(columns=features_to_drop)
y = df[TARGET]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
print(f"Training set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")
print("-" * 50)
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    print(f"--- Training {model_name} ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)    
    accuracy = accuracy_score(y_test, y_pred)
    print(f"\n{model_name} Accuracy Score: {accuracy:.4f}")
    
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))

    cm = confusion_matrix(y_test, y_pred)
    print(f"\n{model_name} Confusion Matrix:")
    print(cm)
    
    print("-" * 50)
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=5)
evaluate_model(knn_classifier, X_train, y_train, X_test, y_test, "K-Nearest Neighbors (KNN)")
```

```python
dt_classifier = DecisionTreeClassifier(random_state=42)
evaluate_model(dt_classifier, X_train, y_train, X_test, y_test, "Decision Tree (DT)")
```

```python
rf_classifier = RandomForestClassifier(n_estimators=50, random_state=42, n_jobs=-1)
evaluate_model(rf_classifier, X_train, y_train, X_test, y_test, "Random Forest (RF)")
```
